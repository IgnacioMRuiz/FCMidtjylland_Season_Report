```python
import pandas as pd
```

```python
# Path to the 'Player Info.xlsx' file and import as df.
players_info_path = "/Users/nacho/Desktop/FCMidtjylland/Data/RawData/Players Info.xlsx"
players_info = pd.read_excel(players_info_path)
```

```python
players_info = players_info.sort_values(by='Full Name')
players_info
```

```python
# Path to the output directory and export of the data frame
output_path = "/Users/nacho/Desktop/FCMidtjylland/Data/ProcessedData/Players Info.xlsx"
players_info.to_excel(output_path, index=False)

print(f"File successfully exported in --> {output_path}")
```
