```python
import pandas as pd
```

```python
# Path to the 'Player Names Master.xlsx' file and import as df.
player_names_master_path = "/Users/nacho/Desktop/FCMidtjylland/Data/RawData/Player Names Master.xlsx"
player_names_master = pd.read_excel(player_names_master_path)
```

```python
# Dictionary creation for mapping full names to their player names
name_mapping = player_names_master.groupby('full_name')['player_name'].apply(list).to_dict()

# Creation of a list to store the rows of the new DataFrame
new_rows = []

# Fill the new_rows list with the transformed data
for full_name, player_names in name_mapping.items():
# It is assumed that there are two ways to call each player. Cases with a single name are still handled.
    new_row = {
        'First Name': player_names[0],
        'First Name Initial': player_names[1] if len(player_names) > 1 else None,  # To handle cases with a single name
        'Full Name': full_name
    }
    new_rows.append(new_row)

# Create the new DataFrame with the new rows
player_names_master = pd.DataFrame(new_rows)
```

```python
player_names_master = player_names_master.sort_values(by='Full Name')
player_names_master
```

```python
# Path to the output directory and export of the data frame
output_path = "/Users/nacho/Desktop/FCMidtjylland/Data/ProcessedData/Player Names Master.xlsx"
player_names_master.to_excel(output_path, index=False)

print(f"File successfully exported in --> {output_path}")
```
