```python
import pandas as pd
```

```python
# Path to the 'Player Info.xlsx' file and import as df.
gps_data_path = "/Users/nacho/Desktop/FCMidtjylland/Data/RawData/GPS Data.xlsx"
gps_data = pd.read_excel(gps_data_path)
```

```python
# Separation of the 'Drill Title' column into new columns
gps_data[['Drill Type', 'Half', 'Time Division', 'Players Involved', 'Size', 'Additional Info']] = gps_data['Drill Title'].str.split(';', expand=True)

# Adjust the column 'Half' and 'Time Division'.
gps_data['Half'] = gps_data['Half'] + " " + gps_data['Time Division']

# Remove excess columns
gps_data.drop(columns=['Time Division', 'Drill Title'], inplace=True)
```

```python
# Replace words
gps_data['Drill Type'] = gps_data['Drill Type'].replace('game', 'Game')
gps_data['Size'] = gps_data['Size'].replace('full pitch', 'Full pitch')
# Rename column name
gps_data.rename(columns={'Session Date': 'Date'}, inplace=True) 
gps_data.rename(columns={'Player Name': 'First Name Initial'}, inplace=True)
```

```python
# Path to the 'Player Info.xlsx' file and import as df.
players_involvements_path = "/Users/nacho/Desktop/FCMidtjylland/Data/ProcessedData/Players Involvements.xlsx"
players_involvements = pd.read_excel(players_involvements_path)
```

```python
# Create a Date to Match mapping dictionary from players_involvements
date_to_match_mapping = players_involvements.drop_duplicates('Date').set_index('Date')['Match'].to_dict()

# Apply the mapping to the DataFrame gps_data to create the new Match column
gps_data['Match'] = gps_data['Date'].map(date_to_match_mapping)
```
